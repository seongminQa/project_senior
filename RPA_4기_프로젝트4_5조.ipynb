```python
# 거주 선호도 -> 노인독거 (혼자 사는 노인 분들은 일반주택이나 아파트에서 홀로 사시기를 원한다. 자녀들과의 동거는 원하지 않음. 눈에띄는 결과는 시설입소를 희망하는 분들이 있다는 것이다.)
ratio = (77.5, 8.5, 1.2, 1.6, 0.7, 10.4)
labels = ('일반주택_아파트','자녀동거','고령자_전용_주거단지','그룹홈','일반주택_아파트+경로당','시설입소')
explode = [0, 0.20, 0.95, 1.10, 1.2, 0.20]
plt.figure(figsize = (10,7))
title_font = {'fontsize': 16, 'fontweight': 'bold'}
plt.title('주거지 선호도(노인독거)')
plt.pie(ratio, labels = labels, autopct = '%.1f%%', startangle=260, counterclock=False, explode = explode, textprops = {'fontsize':16, 'color': 'k'})
plt.show()
```

```python
# 거주 선호도 -> 노인부부 ->(일반주택이나 아파트를 선호한다. 자녀들과 사는 것을 선호하지는 않는다.)
ratio = (88.8, 5.9, 0.7, 0.6, 0.3, 3.6)
labels = ('일반주택_아파트','자녀동거','고령자_전용_주거단지','그룹홈','일반주택_아파트+경로당','시설입소')
explode = [0, 0.20, 0.95, 1.10, 1.2, 0.20]
plt.figure(figsize = (10,7))
plt.title('주거지 선호도(노인부부)')
plt.pie(ratio, labels = labels, autopct = '%.1f%%', startangle=265, counterclock=False, explode = explode, textprops = {'fontsize':16, 'color': 'k'})
plt.show()
```
